```python
chunks = [

    "Employees receive 12 casual leaves annually.",

    "Employees receive 15 sick leaves annually.",

    "Employees may work from home twice per week.",

    "Travel expenses are reimbursed within 30 days.",

    "All employees are covered under company medical insurance."

]

print("Total Chunks:", len(chunks))
```

```python
from sentence_transformers import SentenceTransformer
model=SentenceTransformer('all-MiniLM-L6-v2')
print("Embedding Model Loaded.")
```

```python
emb=model.encode(chunks)
print("Embedding Generated.")
```

```python
print("Embedding Shape:",emb.shape)
```

```python
!pip install faiss-cpu
```

```python
import faiss
dimension=emb.shape[1]
index=faiss.IndexFlatL2(dimension)
print("FAISS Index Generated.")
```

```python
import numpy as np
index.add(np.array(emb,dtype='float32'))
```

```python
print("Vectors Stored:",index.ntotal)
```

```python
query="How many casual leaves are allowed?"
print(query)
```

# Query to Embeddings

```python
query_emb=model.encode(
    [query]
)
```

```python
k=3
distances,indices=index.search(np.array(query_emb,dtype=np.float32),k=k)
print("Distances:",distances)
print("Indices:",indices)
```

```python
print("Matching Chunks\n")
for rank,idx in enumerate(indices[0]):
  print(f'Rank {rank+1}:')
  print(chunks[idx])
  print(f'Distance:',distances[0][rank])
  print("-"*50)
```

## Task 6: Integrate Large Language Model

Now, we'll use a Large Language Model (LLM) to generate natural-language answers based on the retrieved chunks. We will use `Gemma 2B` for this task.

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Load the TinyLlama model and tokenizer
print("Loading TinyLlama model...")
tokenizer = AutoTokenizer.from_pretrained("TinyLlama/TinyLlama-1.1B-Chat-v1.0")
model = AutoModelForCausalLM.from_pretrained("TinyLlama/TinyLlama-1.1B-Chat-v1.0", device_map="auto")
print("TinyLlama model loaded successfully!")
```

```python
def generate_answer(question, retrieved_chunks):
    context = "\n".join(retrieved_chunks)
    prompt = f"""<|system|>
You are a helpful assistant. You will be provided with a question and relevant context. Based on the context, provide a concise and accurate answer. If the answer cannot be found in the context, state that.
<|user|>
Context:
{context}
Question:
{question}
Answer:"""

    input_ids = tokenizer(prompt, return_tensors="pt").to(model.device)

    # Generate output with attention_mask
    outputs = model.generate(input_ids=input_ids.input_ids, attention_mask=input_ids.attention_mask, max_new_tokens=200, temperature=0.7)
    full_generation = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # The model often repeats the input prompt. We need to extract only the generated answer.
    # Find the end of the original prompt in the full generation.
    prompt_end_index = full_generation.rfind(prompt)

    if prompt_end_index != -1:
        # The actual generation starts right after the prompt
        model_raw_response = full_generation[prompt_end_index + len(prompt):].strip()

        # Now, clean up any subsequent questions/turns the model might have accidentally generated.
        # Look for typical delimiters like "Question:", "<|user|>", "<|system|>", "Answer:"
        stop_patterns = ["Question:", "<|user|>", "<|system|>", "Answer:"]

        # Find the first occurrence of any stop pattern in the model's raw response
        first_stop_index = len(model_raw_response)
        for pattern in stop_patterns:
            idx = model_raw_response.find(pattern)
            if idx != -1 and idx < first_stop_index:
                first_stop_index = idx

        final_answer = model_raw_response[:first_stop_index].strip()

    else:
        # Fallback if the original prompt structure isn't found (unlikely with this setup)
        final_answer = full_generation.strip()

    return final_answer
```

```python
print("=" * 50)
print("GENERATING ANSWERS WITH TINYLAMA")
print("=" * 50)

k = 3  # Number of top similar chunks to retrieve

for query_idx, query_text in enumerate(q):
    retrieved_chunks_for_llm = []
    print(f"\n{'#' * 80}")
    print(f"Query: {query_text}")
    print(f"{'#' * 80}")

    # Use the pre-computed indices `i` and distances `d` for this query
    current_indices = i[query_idx]
    current_distances = d[query_idx]

    for rank in range(k):
        chunk_index = current_indices[rank]
        retrieved_chunk_content = chunks[chunk_index] # Use 'chunks' which is defined
        retrieved_chunks_for_llm.append(retrieved_chunk_content)
        print(f"\n--- Retrieved Chunk {rank+1} (from chunks[{chunk_index}]):")
        print(retrieved_chunk_content)
        print(f"  Similarity Score (L2 Distance): {current_distances[rank]:.4f}")

    print(f"\n{'=' * 80}")
    print("Generated Answer:")
    generated_answer = generate_answer(query_text, retrieved_chunks_for_llm)
    print(generated_answer)
    print(f"{'=' * 80}")
```
